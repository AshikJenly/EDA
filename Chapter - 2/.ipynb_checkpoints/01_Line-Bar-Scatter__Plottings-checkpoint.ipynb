# Line Chart

```python
import datetime
import math
import pandas as pd
import random
import radar
from faker import Faker
fake = Faker()
```

```python
def generateData(n):
    listdata = []
    start = datetime.datetime(2019, 8, 1)
    end = datetime.datetime(2019, 8, 30)

    delta = end - start
    for _ in range(n):
        date = radar.random_datetime(start='2019-08-1',
        stop='2019-08-30').strftime("%Y-%m-%d")
        price = round(random.uniform(900, 1000), 4)
        listdata.append([date, price])
    df = pd.DataFrame(listdata, columns = ['Date', 'Price'])
    df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
    df = df.groupby(by='Date').mean()
    return df
```

```python
df = generateData(50)
```

# Bar Chart 

```python
import calendar
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
months = list(range(1, 13))
sold_quantity = [round(random.uniform(100, 200)) for x in range(1,13)]

figure, axis = plt.subplots()
plt.xticks(months, calendar.month_name[1:13], rotation=20)
axis.bar(months,sold_quantity)
plt.show()
```

```python
months = list(range(1, 13))
sold_quantity = [round(random.uniform(100, 200)) for x in range(1,13)]

figure, axis = plt.subplots()
plt.xticks(months, calendar.month_name[1:13], rotation=20)
plot=axis.bar(months,sold_quantity)

for rectangle in plot:
    height = rectangle.get_height()
    axis.text(rectangle.get_x() + rectangle.get_width() /2., 1.003 *height, '%d' % int(height), ha='center', va = 'bottom')
plt.show()
```

```python
for r in plot:
    print('type : ',type(r))
    print('height : ',r.get_height())
    print('X  : ',r.get_x())
    print('width  : ',r.get_width())   
    print('--------------------------------')
```

```python
months = list(range(1, 13))
sold_quantity = [round(random.uniform(100, 200)) for x in range(1, 13)]
figure, axis = plt.subplots()
plt.yticks(months, calendar.month_name[1:13], rotation=20)
plot = axis.barh(months, sold_quantity)
# for rectangle in plot:
#     width = rectangle.get_width()
#     axis.text(width + 2.5, rectangle.get_y() + 0.38, '%d' % int(width),
#     ha='center', va = 'bottom')
plt.show()
```

# Scatter Plot


    Scatter plots are also called scatter graphs, scatter charts, scattergrams, and scatter
    diagrams. They use a Cartesian coordinates system to display values of typically two
    variables for a set of data.



When should we use a scatter plot? Scatter plots can be constructed in the following two
situations:


   <li>When one continuous variable is dependent on another variable, which is under
         the control of the observer</li>
   <li>When both continuous variables are independent </li>

#### Loading required data

```python
headers_cols = ['age','min_recommended', 'max_recommended','may_be_appropriate_min', 'may_be_appropriate_max', 'min_not_recommended','max_not_recommended']
sleepDf =pd.read_csv('https://raw.githubusercontent.com/PacktPublishing/hands-on-exploratory-data-analysis-with-python/master/Chapter%202/sleep_vs_age.csv',index_col=0)
sleepDf.head(10)
```

```python
sleepDf.columns == headers_cols
```

####  import  required modules

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
# A regular scatter plot
plt.scatter(x=sleepDf["age"]/12., y=sleepDf["min_recommended"])
plt.scatter(x=sleepDf["age"]/12., y=sleepDf['max_recommended'])
plt.xlabel('Age of person in Years')
plt.ylabel('Total hours of sleep required')
plt.show()
```

```python
#Line plot of same data

plt.plot(sleepDf['age'],sleepDf['min_recommended'],'r--')
plt.plot(sleepDf['age'],sleepDf['max_recommended'],'g--')
plt.xlabel('Age of person in Years')
plt.ylabel('Total hours of sleep required')
plt.show()
```

## using seaborn to load the dataset:

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (8, 6)
# plt.rcParams['figure.dpi'] = 150
sns.set()
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['species'] = df['species'].map({'setosa': 0, "versicolor": 1,"virginica": 2})
```

```python
df.head()
```

```python
plt.scatter(x=df["sepal_length"], y=df["sepal_width"], c =df.species)

plt.xlabel('Septal Length')
plt.ylabel('Petal length')
plt.show()
```

# Bubble Chart

```python
# Create bubble plot

plt.scatter(df.petal_length, df.petal_width,
            s=50*df.petal_length*df.petal_width,    
            c=df.species,
            alpha=0.3
           )

# Create labels for axises
plt.xlabel('Septal Length')
plt.ylabel('Petal length')
plt.show()
```
