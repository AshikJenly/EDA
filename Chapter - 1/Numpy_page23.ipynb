### styling the data

```python
def colorNegativeValueToRed(value):
    if value < 0:
        color = 'red'
    elif value > 0:
        color = 'black'
    else:
        color = 'green'
    return 'color: %s' % color
```

```python
colorNegativeValueToRed(1.33)
```

```python
dFrame.style.applymap(colorNegativeValueToRed,subset=['A','B','C','D','E'])
```

### Highlight Min and Max values 

```python
abc = np.arange(1,12)
```

```python
abc
```

```python
[e if e!=4 else '' for e in abc if e %2 == 0]
```

```python
def highlight(s):
    isMax = s == s.max()
    
    return ['background-color: orange' if v else '' for v in isMax]
def highlightMin(s):
    isMin = s == s.min()
    return ['background-color: green' if v else '' for v in isMin]
```

```python
dFrame.style.apply(highlight).apply(highlightMin).highlight_null(null_color='red')
```
