## Data Refactoring

    Data refactoring is the process of modifying the organization and structure of data in a database to optimize its quality, efficiency, and usability.

```python
import re
```

```python
string ="'Coursera' <no-reply@m.mail.coursera.org>"
```

```python
re.findall(r'<(.+?)>', string)
```

```python
list(filter(lambda y: '@' in y, string.split()))
```

```python
def extract_email_ID(string):
    email = re.findall(r'<(.+?)>', string)
    if not email:
        email = list(filter(lambda y: '@' in y, string.split()))
    return email[0] if email else np.nan
```

```python
df['from'] = df['from'].apply(lambda x: extract_email_ID(x))
```

```python
myemail = 'ashikjenoly@gmail.com'
df['label'] = df['from'].apply(lambda x: 'sent' if x==myemail else 'inbox')
```
